```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import mdtraj as md
```

```python
hills1 = pd.read_csv('HILLS_omega1', names=["time", "omega1", "sigma_omega1", "height", "biasf"], comment="#", delim_whitespace=True)
hills2 = pd.read_csv('HILLS_omega2', names=["time", "omega2", "sigma_omega2", "height", "biasf"], comment="#", delim_whitespace=True)
hills3 = pd.read_csv('HILLS_omega3', names=["time", "omega3", "sigma_omega3", "height", "biasf"], comment="#", delim_whitespace=True)
biasfactor = 30/29
```

```python
plt.plot(hills1.time/1000, hills1.height)
plt.axhline(y=(hills1[round(len(hills1)*-0.1):].height).mean(), color='green', linestyle='-')
plt.axhline(y=(biasfactor)*1.2*0.1, color='black', linestyle='-')
plt.axhline(y=(biasfactor)*1.2*0.05, color='red', linestyle='-')
plt.xlabel("Simulation Time (ns)")
plt.ylabel("Deposited Gaussian Heights (kJ/mol)")
plt.title("PC0 Gaussians deposited over time")
print("average over terminal 10 percent of simulation: %f" %(hills1[round(len(hills1)*-0.1):].height).mean())
print("10 percent initial height: %f" %((biasfactor)*1.2*0.1))
print("5 percent initial height: %f" %((biasfactor)*1.2*0.05))
#plt.plot(r1_pc2.time/1000, r1_pc2.pc1)
#plt.axhline(y=1.2*0.05, color='black', linestyle='-')
#plt.xlabel("Simulation Time (ns)")
#plt.ylabel("PC1")
#plt.title("PC1 over time")
#plt.hist(r1_pc1.pc1, bins=100)
```
